### Import all the libraries

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

### Set the based link

```python
base_url = 'https://www.jumia.com.ng/'
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36"
}
```

### Get the Unique Link for each product

```python
pagelinks = []
for x in range(1,31):
    r = requests.get(f"https://www.jumia.com.ng/catalog/?q=shoes&page={x}#catalog-listing")
    soup = BeautifulSoup(r.content, 'html.parser')
    productcatalog = soup.find_all('div', class_= "-paxs row _no-g _4cl-3cm-shs")
    
    for item in productcatalog:
        for link in item.find_all('a', href = True):
            pagelinks.append(base_url+link['href'])
print(pagelinks)
```

```python
MenShoeList = []
for link in pagelinks:
    r = requests.get(link, headers = headers)
    soup = BeautifulSoup(r.content, 'html.parser')

    try:
        shoe_name = soup.find('h1', class_="-fs20 -pts -pbxs").text.strip()
        price = soup.find('span', class_='-b -ltr -tal -fs24 -prxs').text.strip()[1:]
        ratings = soup.find('div', class_='stars _m _al').text.strip()[:3]
        reviews = soup.find('a', class_='-plxs _more').text.strip()[1:3]
    
    except TypeError:
        print(None)
    
    MenShoes = {
        
        'Shoe Name': shoe_name,
        'Price': price,
        'Ratings': ratings,
        'Reviews': reviews        
        }
    
    MenShoeList.append(MenShoes)
    print('Saving:', MenShoes['Shoe Name'])
```

```python
df = pd.DataFrame(MenShoeList)
```
